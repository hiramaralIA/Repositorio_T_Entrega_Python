```python
# constantes: JUROS_POUP;
# Variáveis: valorDepositado, valorComRedimento ;
# Operadores matemáticos: + e *;
# Operadore de atribuição: = ;
# Operadore de concatenação: , ;

JUROS_POUP = 0.70
valorDepositado = float(input("Informe o valor ser depositado: "))

valorComRedimento = valorDepositado + (valorDepositado * JUROS_POUP)

print("Valor com rendimento: ", valorComRedimento)
```
